```python
dtipos = {
    'ID': str,
    'entregador_ID': str,
    'entregador_idade': object,
    'avaliacao_entregador': float,
    'restaurante_lat': float,
    'restaurante_long': float,
    'entrega_lat': float,
    'entrega_long': float,
    'data_pedido': str,
    'hora_pedido': str,
    'hora_entrega': str,
    'clima': str,
    'densidade_trafego': str,
    'condicao_veiculo': object,
    'tipo_pedido': str,
    'tipo_veiculo': str,
    'varias_entregas': object,
    'feriado': str,
    'regiao': str,
    'tempo_entrega': str
}
```

```python
df.loc[df['feriado']=='No', 'feriado'] = False
df.loc[df['feriado']=='Yes', 'feriado'] = True
```

```python
df['data_pedido'].unique()
```

```python
df['data_hora_pedido'] = pd.to_datetime(df['data_pedido']+ ' ' + df['hora_pedido'], dayfirst=True)
```

```python
df['data_hora_entrega'] = pd.to_datetime(df['data_pedido']+ ' ' + df['hora_entrega'], dayfirst=True)
```

```python
df.drop(columns=['data_pedido', 'hora_pedido', 'hora_entrega'], inplace=True)
```

```python
df['tempo_entrega'] = df['tempo_entrega'].str.replace('(min) ', '')
```

```python
df['tempo_entrega'] = df['tempo_entrega'].astype(int)
```

```python
df['clima'] = df['clima'].str.replace('conditions ', '')
```

```python
df['clima']
```

```python
def one_hot_encode(dataframe, categorical_cols):
  for cat_col in categorical_cols:
    cat_col_values = dataframe[cat_col].unique()
    cat_col_values.sort()
    for col in cat_col_values:
      dataframe[col] = dataframe[cat_col].apply(lambda x: x==col).astype(int)
```

```python
df
```

```python
cat_cols = ['clima', 'densidade_trafego', 'tipo_pedido', 'tipo_veiculo', 'regiao']
one_hot_encode(df, cat_cols)
```

```python
df.drop(columns=cat_cols, inplace=True)
```
